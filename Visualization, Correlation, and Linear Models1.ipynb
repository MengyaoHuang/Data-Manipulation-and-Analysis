### Some supplemental information about the Wheel itself:
* Abstraction and Figuration : Level of figuration in a graphic depends on the phenomenon being depicted using physical representations, such as photographs or drawings. As the representations shift from real phenomenon to less real and more conceptual, the emphasis shifts from figuration to abstraction.
* Functionality and Decoration : A completely functional graphic has no decorations and is likely to be a direct representation of the data. On the other hand, a highly decorated graphic turns out to be artistically embroidered. These decorations may increase the amount of time a viewer spends considering the visual but sometimes exploring those nuances form mental associations thereby increasing familiarity and memorability.
* Density and Lightness : They relate to the amount of information being portrayed. This dimension comes to play when the reader(or audience) is heavily content engaged i.e. density is important unlike the quick narrative ones who prefer lightness i.e. more overview information rather than depth.
* Multidimensional and Uni-dimensional : A multidimensional graphic explains a phenomena considering more number of dimensions allowing viewer to explore different aspects. A uni-dimensional graphic instead focuses on a single or a few dimensions i.e. tries to explore the phenomena in limited ways.
* Originality and Familiarity : Among the different kinds of information graphics like bar charts, line charts, scatter plots and many more. Thinking requires here is in terms of these representations fit well for representing which information of the phenomena. For certain cases one type of graphic say scatter plots are better than bar charts or vice versa. To judge this, the originality and familiarity trade off matters a lot.
* Novelty and Redundancy : Redundancy in a graphic is showing the same information in many different ways. For example, using bar charts and putting colours for different bars depicting which one is higher. Here, in this example the colour is redundant as there’s no use of it. This make your graphics highly complex and confuse the readers. Therefore, novelty is the act of describing each information in the graphic in only one way.

## T-test

A t-test is a simple statistical model that's commonly used to test whether the means of two different
distributions are the same.  scipy.stats gives us a handy interface for this:

````python
goodwines = wine.query('isgood == "good"')
badwines = wine.query('isgood == "bad"')
stats.ttest_ind(goodwines["fixed acidity"], badwines["fixed acidity"])
````

````python
stats.ttest_ind(wine[wine.isgood == 'good']['fixed acidity'],wine[wine.isgood == 'bad']['fixed acidity'])
````

## Ordinary Least Squares (OLS) Regression

We can get a lot more detail about the regression model by using statsmodels

````python
import statsmodels.api as sm
import statsmodels.formula.api as smf
````

statsmodels uses R-Style formula: y ~ x1 + x2 + x3 + ...

1. y represents the outcome/dependent variable
2. x1, x2, x3, etc represent explanatory/independent variables 

````python
model1 = smf.ols('chlorides ~ sulphates', data=wine).fit()
model1.summary()
````

### Interesting things happen when we use OLS to do an ANOVA (look closely at the model):

````python
model2 = smf.ols('chlorides ~ isgood', data=wine).fit()
model2.summary()
````

### We might want to experiment with the original ```quality``` variable, either in a regression model:

````python
model3 = smf.ols('chlorides ~ quality', data=wine).fit()
model3.summary()
````

### or in an ANOVA (again, look closely at the model):

If quality had been an integer variable that we wanted to treat explicitly as categorical, we could have done so by using the C() operator:

````python
model4 = smf.ols('chlorides ~ C(quality)', data=wine).fit()
model4.summary()
````

## <font color="magenta">Q7: Use OLS to perform either a regression or an ANOVA on a variable (other than chlorides) and interpret your results.

````python
model1 = smf.ols('Q("citric acid") ~ sulphates', data=wine).fit()
model1.summary()
````

````python
model1 = smf.ols('sulphates ~ C(quality)', data=wine).fit()
model1.summary()
````

Comment: 
* We use OLS to perform a regression model based on citric acid and sulphates. This model's R-squared is 0.098, which indicates that the model fails to fit the data very well. Coefficients of intercept and sulphates are 0.0344 and 0.3594 and both their p-values indicate that coefficients are significant in this model. 
* And we use ANOVA on the variable 'sulphates'. This model's R-squared is 0.065, which indicates that the model fails to fit the data very well. By default, this model treats quality categorical variables with all possible values as dummy boolean variables and take the last level into intercept term. According to the output, the coefficients of intercept and quality 6, 7, 8 are significant here in predicting sulphates under OLS model.

We've provided the CSV file for this lab so you can go ahead and load it in the usual way:

As always, you should take a look at the data to get a sense of

As we mentioned in an earlier class, pandas is really good at helping
us deal with dates.  The 'Date' column in the dataframe contains 
strings that look like dates.  We can use the ```pandas.to_datetime()``` function to convert the strings to an internal datetime object
(see https://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_datetime.html for more details):

````python
crashes['Date'] = pd.to_datetime(crashes['Date'])
````

And let's look at the dataframe again.  See any difference?

````python
crashes.head(1)
````

The pandas datetime object makes it easy to extract interesting 
parts of the date or time.  In our case, we're interested in extracting
the year, so we can do that with the following code:

````python
crashes['year'] = crashes['Date'].dt.year
crashes.head()
````

And, as always, let's look at what we got:

````python
crashes.head(1)
````

As part of the final exercise in this class, let's create a 
visualization of the number of Fatalities per year:

````python
sns.barplot('year','Fatalities',data=crashes)
````
